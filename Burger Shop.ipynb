```python
# Author: Chibuko Ozobu
# Date: 4/25/22
# Assignment: You have been tasked with building a user interface
# for a new restaurant named Burger Shop that allows customers to place custom orders for hamburgers.
# Start by asking the customer for their name, which will be used to identify the order. 
# The program should then perform the following steps:
# Ask the customer if they want a burger, a side, a drink, or a combo.
# A combo must include a burger, a side, and a drink.
# Prompt them for details about their selection, such as condiments for a burger, what kind and size of drink, and so on.
# Create the item based on their selections.
# Add the item to the Order class.
# Repeat these steps until the customer doesn't want to order anything else.
# Display the order details including the price.
# Thank the customer for their business.
# Also include an option that allows the customer to cancel their order at any point in the ordering process. 
# The output should thank the user for their business but not display anything ordered.

# given starter code
class FoodItem:
    price = 0.0
    #currency = " "
    

class Burger(FoodItem):
    burger_type = " " #chicken, beef, turkey, veggie
    toppings = list() #tomatoes, onions, lettuce, pickles, bacon, ketchup, mustard, ranch
    def __init__(self, b_type, t_list):
        self.burger_type = b_type
        if(b_type == "turkey"):
            self.price = 2.50
        if(b_type == "beef"):
            self.price = 3.00
        if(b_type == "veggie"):
            self.price = 3.00
        if(b_type == "chicken"):
            self.price = 2.50
        if len(t_list) == 0:
            self.toppings.append("nothing")
        else:
            self.toppings = t_list
            #self.toppings.pop()  #get rid of the 'no' that is placed in list when saying no more toppings
            for x in self.toppings:
                self.price = self.price + .50 #increase the price of the burger for every topping added         
    def display(self):
        if len(self.toppings) == 0:
            self.toppings.append("nothing")
        print(f"- a ${self.price} {self.burger_type} burger with {self.toppings} on it.")
        
class Drink(FoodItem):
    size = " "
    bev = " "
    def __init__(self, beverage, siz):
        self.size = siz
        if(siz == "small"):
            self.price = 1.50
        if(siz == "medium"):
            self.price = 2.00
        if(siz == "large"):
            self.price = 2.50
        if("extra" in siz):
            self.price = 3.00
        self.bev = beverage
    def display(self):
        print(f"- a ${self.price} {self.size} {self.bev} to drink.")
    
class Side(FoodItem):
    side = " "
    def __init__(self, sid):
        self.side = sid
        if("salad" in sid):
            self.price = 2.5
        elif("fries" in sid):
            self.price = 1.5
        else:
            self.price = 3
    def display(self):
        print(f"- a ${self.price} side of {self.side}.")
    
class Combo(FoodItem):
    brgr = None
    drnk = None
    sde = None
    def __init__(self, b, d, s):
        self.brgr = b
        self.drnk = d
        self.sde = s
        self.price = b.price + d.price + s.price
    def display(self):
        print(f"On the combo you asked for \n---------")
        self.brgr.display()
        self.drnk.display()
        self.sde.display()
        print("---------")
        print("Because you ordered it as a combo you will get $1.50 off your order.")

class Order:
    order = list()
    #dynamically add new FoodItem objects to list
    def __init__(self):
        pass
    def addToOrder(self, food_item):
        self.order.append(food_item)
    def removeOrder(self, food_item):
        self.order.remove(self, food_item)
    def isEmpty(self):
        if len(self.order) == 0:
           return True
        else:
            return False
    def displayOrder(self):
        if len(self.order) == 0:
            print("Error, there is no order")
        else:
            for x in self.order:
                x.display()
    def getPrice(self):
        price = 0.0
        for x in self.order:
            price = price + x.price
            if(isinstance(x,Combo)):
                price = price - 1.5
        return price
    
 
def user_input_burger():
    #ask user for input and store it in burger object 
    kind = input("What kind of burger would you like? We have the options: Turkey, Chicken, Veggie, or Beef.\n").lower()
    toppings = input("Would you like anything on that burger? Any additional toppings add an extra $0.50: ").lower()
    t_list = list()
    while (toppings != "no"):
        t_list.append(toppings)
        toppings = input("Any other toppings? Any additional toppings add an $0.50: ").lower()
    b = Burger(kind, t_list)
    return b
 
def user_input_drink():
    #ask user for input and store it in drink object 
    kind = input("What kind of drink would you like? ").lower()
    size = input("What size drink would you like? (small, medium, large, extra)").lower()
    d = Drink(kind,size)
    return d
 
def user_input_side():
    #ask user for input and store it in side object 
    sd = input("What side would you like? (onion rings, french fries, any salad) ").lower()
    s = Side(sd)
    return s
 
def user_input_combo():
    #ask user for input and store it in combo object 
    #a combo must include one burger, one side, and one drink
    print("For The Combo:")
    cb = user_input_burger() #combo burger
    cd = user_input_drink() #combo drink
    cs = user_input_side() #combo side
    c = Combo(cb, cd, cs)
    return c
 
def take_order():
    final_order = Order()
    final_price = 0.0
    #ask user for name for the order
    print("Hello! Welcome to the MTHREE Burger Shop!\n")
    #print the menu, for every COMBO you get 2 dollars off the order.
    print("\tHere is the menu for the MTHREE Burger Shop \
    \nBurgers\n\t Turkey: $2.50\n\t Chicken: $2.50\n\t Beef: $3.00\n\t Veggie: $3.00 \
    \n\t\t---------- \
    \nAny Drinks\n\t Small: $1.50\n\t Medium: $2.00\n\t Large: $2.50\n\t Extra Large: $3.00\n\t \
    \n\t\t---------- \
    \nSides\n\t Any Salad: $2.50\n\t French Fries: $1.50\n\t Anything Else: $3.00 \
    \n\t\t---------- \
    \n**Combos including a burger, drink, and side and remove $1.50 off your final price**\n") 
    
    customer = input("Can I get a name for the order? ")
    
    order_over = False #bool for if the order is over or not
    c_order = input(f"Okay {customer}, would you like to a BURGER, SIDE, DRINK, or COMBO: ").lower()
    #repeat taking order until client is done 
    while(not order_over):
        if(c_order == "burger"):
            c_b = user_input_burger()
            final_order.addToOrder(c_b)
        if(c_order == "drink"):
            c_d = user_input_drink()
            final_order.addToOrder(c_d)
        if(c_order == "side"):
            c_s = user_input_side()
            final_order.addToOrder(c_s)
        if(c_order == "combo"):
            c_c = user_input_combo()
            final_order.addToOrder(c_c)
            final_price = final_price - 1.50
        c_order = input("Would you like to add a BURGER, SIDE, DRINK, or COMBO? Or are you DONE ordering?\
 Or would you like to CANCEL your order:\n").lower()
        if(c_order == "done"):
            order_over = True
        if("cancel" in c_order):
            order_over = True
        
    if("cancel" in c_order):
        print(f"Sorry that you had to cancel your order {customer}. Have a good day nonetheless.")
    #display order details
    elif(final_order.isEmpty()): #user accidentally finished order too early
        print(f"It seems you finished your order without entering anything {customer}. \
If this was done on purpose have a good day.")
    else:
        print("Your final order was:")
        final_order.displayOrder()
        final_price = final_order.getPrice()
        #display a thank you message
        print(f"The final price for your order was ${final_price}. Thank you {customer} for choosing MTHREE Burger Shop for your meal!")
    
take_order()
```
