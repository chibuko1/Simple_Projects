```python
# Author: Chibuko Ozobu
# Date: 4/22/2022
# Assignment Description: 
# Convert a string to lowercase characters.
# Split the lowercase string into individual words.
# Count the number of words in the lowercase string.
# Determine the number of distinct words in the lowercase string.
# Calculate the number of times each word appears in the lowercase string.
# Remove the punctuation from the lowercase string.
# Perform a count analysis on the text without punctuation characters.

import string

# converting string to lowercase letters
s = """Imagine a vast sheet of paper on which straight Lines, Triangles, Squares, Pentagons, Hexagons, and other figures, instead of remaining fixed in their places, move freely about, on or in the surface, but without the power of rising above or sinking below it, very much like shadows - only hard and with luminous edges - and you will then have a pretty correct notion of my country and countrymen. Alas, a few years ago, I should have said "my universe": but now my mind has been opened to higher views of things."""
s_lower = s.lower()
#print(lowercase)

# splitting the string into words
words = s_lower.split(" ")
#print(f"Here is a list of the words from the string:\n {words}")

# number of lowercase words in original string
num_words = len(words)
#print(f"This is the original number of words (including punctuation marks) in the string: {num_words}")
# count unique words in string
num_unique = len(set(words)) #sets only allow unique values
#print(f"This is the number of unique words (including punctuation marks) in the string: {num_unique}")

# count number of times a word appears
word_freq = dict()

for word in words:
    if(word in word_freq): #if the word already exists in the dictionary
        # increment key's value
        word_freq[word] = word_freq[word] + 1
    else:
        #add key to dict w/ value of 1
        word_freq.update({word:1})
    #print(word)

#print(f"Here is a number of words {word_freq}")

# removing punctuation
punc = list(string.punctuation)
#print(punc)
for word in words:
    for p in punc:
        if word == p:
            words.remove(p)           
#print(f"Here is the new len of words {(len(set(words)))} ")


#everything in a few lines
done = list(s.lower().split(" ")) #list of words w/ punctuations
done_freq = dict()
final_words = list() #list of words without punctuations

for word in done:
    for p in punc:
        if (p == word[len(word)-1] and len(word) > 2 and word != ""): #remove endline punctuation
            word = word[0:len(word)-1]
        elif (p == word[0] and len(word) > 2): #remove frontline punctuation
            word = word[1:len(word)]    
    final_words.append(word)

for word in final_words:
    for p in punc:
        if word == p: #if a word stored in the list is a punctuation, remove it from the list
            final_words.remove(word)
            
for word in final_words:      
    if(word in done_freq): #if the word already exists in the dictionary
        done_freq[word] = word_freq[word] + 1  #increment key's value
    else:
        done_freq.update({word:1}) #add key to dict w/ value of 1
            
print(f"Here is a list of all the words in the string:\n {final_words} \n")
print(f"Here is the number of words in the string: {len(set(final_words))}\n")
print(f"This is the number of times a word appears in the string: \n{done_freq}")
```
